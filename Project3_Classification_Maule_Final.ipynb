```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#importing dataset
cover_df = pd.read_csv('covtype.data',sep=',',header=None,error_bad_lines=False)

#attaching column names
columns = ['elevation','aspect','slope','horizontal_distance_to_hydrology','vertical_distance_to_hydrology','horizontal_distance_to_roadways','hillshade_9am','hillshade_noon','hillshade_3pm','horizontal_distance_to_fire_points','wilderness_area_rawah','wilderness_area_neota','wilderness_area_comanche','wilderness_area_cache','soil_cathedral','soil_vanet','soil_haploborolis','soil_ratake','soil_vanet','soil_vanet_wetmore','soil_gothic','soil_supervisor','soil_troutville', 'soil_bullwark_outcrop','soil_bullwark_land','soil_legault','soil_catamount','soil_pachic','soil_unspecified','soil_cryaquolis','soil_gateview','soil_rogert','soil_typic_cryaquolis','soil_typic_cryaquepts','soil_cryaquolls','soil_leighcan_bouldery','soil_leighcan_substratum_typic','soil_leighcan_stony','soil_leighcan_warm','soil_granile','soil_leighcan_warm_outcrop','soil_leighcan_outcrop','soil_como_legault','soil_como','soil_leighcan_catamount','soil_catamount_outcrop','soil_leighcan_catamount_rock_outcrop','soil_cryorthents','soil_cryumpbrepts','soil_bross','soil_outcrop_cryumbrepts','soil_leighcan_moran','soil_moran_crorthents_leighcan','soil_moran_rock_land','cover_type']
cover_df.columns = columns

#adding a new feature based on the dataset
cover_df['euc_distance_to_hydrology'] = np.sqrt(cover_df['horizontal_distance_to_hydrology']**2 * cover_df['vertical_distance_to_hydrology']**2)

#converting our class label, cover_type to a string rather than a numerical variable
cover_df['cover_type'] = cover_df['cover_type'].astype('str')

cover_df.head()
```

```python
np.shape(cover_df)
```

```python
cover_df.info()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#normalize each quantitative variable, X ~ N(0, 1)
for val in ['elevation','aspect','slope','horizontal_distance_to_hydrology','euc_distance_to_hydrology','vertical_distance_to_hydrology','horizontal_distance_to_roadways','hillshade_9am','hillshade_noon','hillshade_3pm','horizontal_distance_to_fire_points']:
    print(val)
    x_scaled = scaler.fit_transform(X=cover_df[val].values.reshape(-1, 1))
    cover_df[val] =  x_scaled
    
cover_df.head()
```

```python
cover_df.describe()
```

```python
import matplotlib
params = {'axes.titlesize':'30',
          'xtick.labelsize':'20',
          'ytick.labelsize':'20'}
matplotlib.rcParams.update(params)

#plotting historgrams of all variables
%matplotlib inline
cover_df.iloc[:,0:19].hist(bins=50, figsize=(40,40))
plt.show()
```

```python
#plotting historgrams of all variables
%matplotlib inline
cover_df.iloc[:,20:54].hist(bins=50, figsize=(40,40))
plt.show()
```

```python
from sklearn.model_selection import train_test_split

#splitting data into training, testing sets
train_set, test_set = train_test_split(cover_df, test_size=0.8, random_state=0)

#splitting training set into labels, other data
train_set_Y = train_set['cover_type']
train_set_X = train_set.drop(columns=['cover_type'])
train_set_X.head()
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(max_iter=5, tol=-np.infty, random_state=42)
sgd_clf.fit(train_set_X, train_set_Y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, train_set_X, train_set_Y
                , cv=8, scoring="accuracy")
```

```python
np.mean(cross_val_score(sgd_clf, train_set_X, train_set_Y
                , cv=8, scoring="accuracy"))
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(sgd_clf, train_set_X, train_set_Y, cv=8)
print('Precision: ',precision_score(train_set_Y, y_train_pred, average="macro"))
print('Recall: ', recall_score(train_set_Y, y_train_pred, average="macro"))
```

```python
def model_scoring(model, train_set_X, train_set_Y):
    """
    Print accuracy score, precision, recall for a given model with given training data and validation labels
    """
    print('Accuracy Score: ',np.mean(cross_val_score(model, train_set_X, train_set_Y, cv=8, scoring="accuracy")))
    y_train_pred = cross_val_predict(model, train_set_X, train_set_Y, cv=8)
    print('Precision: ',precision_score(train_set_Y, y_train_pred, average="macro"))
    print('Recall: ', recall_score(train_set_Y, y_train_pred, average="macro"))
```

```python
from sklearn.metrics import f1_score

f1_score(train_set_Y, y_train_pred, average="weighted")
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(train_set_Y, y_train_pred)
conf_matrix
```

```python
np.fill_diagonal(conf_matrix, 0)

plt.matshow(conf_matrix, cmap=plt.cm.Purples)
plt.show()
```

```python
row_sums = conf_matrix.sum(axis=1, keepdims=True)
norm_conf_matrix = np.log(conf_matrix / row_sums)
```

```python
np.fill_diagonal(norm_conf_matrix, 0)

plt.matshow(norm_conf_matrix, cmap=plt.cm.Purples)
plt.show()
```

# KNearest Neighbors + GridSearch

```python
from sklearn.model_selection import GridSearchCV

#using grid search to find the best hyperparameters for our decision tree model
param_grid = [
    # try 45 (5×9) combinations of hyperparameters
    {'n_neighbors': [3,5,7,9]},
  ]

knn_grid = KNeighborsClassifier()
# train across 5 folds, that's a total of (63)*5=315 rounds of training 
grid_search = GridSearchCV(knn_grid, param_grid, cv=5,
                           scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(train_set_X, train_set_Y)
```

```python
grid_search.best_params_
```

```python
(pd.DataFrame(grid_search.cv_results_).sort_values(by=['mean_test_score'],ascending=False)).head()
```

```python
y_train_best_knn_pred = cross_val_predict(best_knn, train_set_X, train_set_Y, cv=8, n_jobs=-1)
f1_score(train_set_Y, y_train_best_knn_pred, average="weighted")
```

```python
model_scoring(best_knn, train_set_X, train_set_Y)
```
